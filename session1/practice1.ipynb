# BAG OF WORDS

```python
corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"]
vocab = []
tokenized = [sentence.split() for sentence in corpus]
# print(tokenized)
for x in tokenized:
    for y in x:
        if y not in vocab:
            vocab.append(y)
vocab = sorted(set(vocab))
# print(vocab)

a = input("Enter a string: ")
asplit = a.split()

bow = {}

for word in vocab:
    bow[word] = asplit.count(word)

valbow = list(bow.values())
valbow.sort()
print(valbow)
```

```python
import re

string = 'hello 12 12 hi 89. Howdy 34'
pattern = r'\d'

results = re.findall(pattern, string)
print(results)
```

```python
import re

str = 'My email address are example@example.com and test@test.edu.vn'
pattern = r'[a-zA-Z0-9.]+@[a-zA-Z0-9.-]+\.[a-zA-Z]{3,}'

results = re.findall(pattern, str)
print(results)
```
