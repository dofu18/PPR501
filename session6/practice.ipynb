```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('uber-raw-data-sep14.csv')
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
# df['Day'] = df['Date/Time'].dt.day
df["Day"] = df['Date/Time'].apply(lambda x: x.day)
df["Hour"] = df['Date/Time'].apply(lambda x: x.hour)
df["Weekday"] = df['Date/Time'].apply(lambda x: x.weekday())

# print(df)

sns.histplot(data=df, x="Day", stat='density', kde=True)
```

```python
sns.histplot(data=df, x="Hour", stat='density', kde=True)
```

```python
sns.histplot(data=df, x="Weekday", stat='density', kde=True)
```

```python
#Correlation of Weekday and Hour heatmap
heatmap_data = df.groupby(['Weekday', 'Hour']).size().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data)
plt.title('Heatmap of Uber Rides by Weekday and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.show()
```

```python
#Scatter: Lat vs. Lon
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x="Lon", y="Lat")
plt.title('Geographical Distribution of Uber Rides')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
