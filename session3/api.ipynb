```python

from flask import Flask
import pandas as pd
#import json

app = Flask(__name__)

@app.route('/churn__count')
def churn_count():    # Load the dataset
    df = pd.read_csv('Telco-Customer-Churn.csv')
    result = df['Churn'].value_counts().to_dict()
    return result

if __name__ == '__main__':
    app.run()
```

```python
from flask import Flask, request

app = Flask(__name__)

@app.route('/get_customer_info')
def get_customer_info():
    customerID = request.args.get('customerID')

    print(customerID)
    df = pd.read_csv('Telco-Customer-Churn.csv')
    result = df[df['customerID'] == customerID].to_dict()
    return result

if __name__ == '__main__':
    app.run()
```
