```python
import pandas as pd

df = pd.read_csv('studentlist.csv')
# print(df)
```

```python
#Thống kê lớp có bao nhiêu bạn có điểm từ 8 trở lên
# count = 0
# for index, row in df.iterrows():
#     if row['Diem'] >= 8:
#         count += 1
        
count = df[df['Diem'] >= 8].count()['Diem']

# print(count.shape)

print("Số bạn có điểm từ 8 trở lên:", count)
```

```python
# Thống kê lớp có bao nhiêu bạn điểm dưới 4 hoặc không có điểm
# count_low = 0
# for index, row in df.iterrows():
#     if row['Diem'] < 4 or pd.isna(row['Diem']):
#         count_low += 1

count_low = df[(df['Diem'] < 4) | (df['Diem'].isna())].shape[0]

print("Số bạn có điểm dưới 4 hoặc không có điểm:", count_low)
```
