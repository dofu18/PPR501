```python
def calculate_can_chi_calendar(year: int) -> str:
    can_list = [
        "Canh", "Tân", "Nhâm", "Quý",
        "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"
    ]

    chi_list = [
        "Thân", "Dậu", "Tuất", "Hợi",
        "Tý", "Sửu", "Dần", "Mẹo",
        "Thìn", "Tỵ", "Ngọ", "Mùi"
    ]

    can = can_list[year % 10]
    chi = chi_list[year % 12]

    return f"{can} {chi}"


if __name__ == "__main__":
    year_input = int(input("Enter a year: "))
    result = calculate_can_chi_calendar(year_input)
    print(f"The Can Chi for the year {year_input} is: {result}")
```

```python
def calculate_investment_value(P, PMT, yearly_rates):
    balance = P

    for rate in yearly_rates:
        monthly_rate = rate / 12
        for _ in range(12):
            balance *= (1 + monthly_rate)  # tính lãi tháng
            balance += PMT                 # góp tiền cuối tháng

    return balance


if __name__ == "__main__":
    principal = float(input("Nhập số tiền ban đầu (P): "))
    monthly_payment = float(input("Nhập số tiền góp hàng tháng (PMT): "))
    years = int(input("Nhập số năm: "))

    yearly_interest_rates = []
    for year in range(1, years + 1):
        rate = float(input(f"Nhập lãi suất hàng năm cho năm {year} (dưới dạng số thập phân): "))
        yearly_interest_rates.append(rate)

    final_value = calculate_investment_value(principal, monthly_payment, yearly_interest_rates)
    print(f"Giá trị đầu tư cuối cùng sau {years} năm là: {final_value:.2f}")
```

```python
import random

def random_number_with_condition():
    count = 0

    while True:
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        count += 1

        if a + b == 40:
            break

    return count

if __name__ == "__main__":
    attempts = random_number_with_condition()
    print(f"Số lần thử trước khi tổng bằng 40: {attempts}")
```

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack = []

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.capacity

    def push(self, value):
        if self.isFull():
            raise Exception("Stack is full")
        self.stack.append(value)

    def pop(self):
        if self.isEmpty():
            raise Exception("Stack is empty")
        return self.stack.pop()

    def top(self):
        if self.isEmpty():
            raise Exception("Stack is empty")
        return self.stack[-1]


stack1 = Stack(capacity=5)

stack1.push(1)
stack1.push(2)
stack1.push(3)
stack1.push(4)
stack1.push(5)

print(stack1.isFull())   # True
print(stack1.top())      # 5
print(stack1.pop())      # 5
print(stack1.top())      # 4
print(stack1.pop())      # 4
print(stack1.top())      # 3
print(stack1.isEmpty())  # False
```

```python
from flask import Flask, jsonify, request

app = Flask(__name__)
app.json.sort_keys = False

# Danh sách sách (giả lập database)
books = []
next_id = 1
@app.route("/books", methods=["GET"])
def get_books():
    return jsonify(books)
@app.route("/books", methods=["POST"])
def add_book():
    global next_id
    data = request.get_json()

    book = {
        "id": next_id,
        "title": data["title"],
        "author": data["author"]
    }

    books.append(book)
    next_id += 1

    return jsonify(book), 201
@app.route("/books/<int:book_id>", methods=["PUT"])
def update_book(book_id):
    data = request.get_json()

    for book in books:
        if book["id"] == book_id:
            book["title"] = data.get("title", book["title"])
            book["author"] = data.get("author", book["author"])
            return jsonify(book)

    return jsonify({"error": "Book not found"}), 404
@app.route("/books/<int:book_id>", methods=["DELETE"])
def delete_book(book_id):
    for book in books:
        if book["id"] == book_id:
            books.remove(book)
            return jsonify({"message": "Book deleted"})

    return jsonify({"error": "Book not found"}), 404


if __name__ == "__main__":
    app.run(host="127.0.0.1", port=5000, debug=False)
```
