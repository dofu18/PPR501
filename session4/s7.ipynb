```python
import sqlite3
import pandas as pd

connection = sqlite3.connect('database.db')
# cursor = connection.cursor()
print(connection)
```

```python
sql_query = """
create table customers (
    email text primary key,
    name text not null,
    phone numeric unique
);
"""
cursor = connection.cursor()
cursor.execute(sql_query)
connection.commit()
print("Table created successfully")
```

```python
sql_query = """
insert into customers (
    email,
    name,
    phone
) values (
    'phudht@gmail.com',
    'Do Phu',
    1234567890
);
"""
cursor = connection.cursor()
cursor.execute(sql_query)
connection.commit()
print("Table inserted successfully")
```

```python
data = pd.read_sql_query("SELECT * FROM customers;", connection)
print(data)
```

```python
values = [
    ('phu1@gmail.com', 'Phu One', 1111111111),
    ('phu2@gmail.com', 'Phu Two', 2222222222),
    ('phu3@gmail.com', 'Phu Three', 3333333333)
]

sql_query = """
insert into customers (email, name, phone) values (?, ?, ?);
"""

cursor = connection.cursor()
cursor.executemany(sql_query, values)
connection.commit()
print("Multiple records inserted successfully")

data = pd.read_sql_query("SELECT * FROM customers;", connection)
print(data)
```

```python
sql_query = """
update customers
set phone = 9999999999
where email = 'phu1@gmail.com'
;
"""
cursor = connection.cursor()
cursor.execute(sql_query)
connection.commit()
print("Record updated successfully")
```

```python
sql_query = """
delete from customers
where email = 'phu1@gmail.com'
;
"""
cursor = connection.cursor()
cursor.execute(sql_query)
connection.commit()
print("Record deleted successfully")

data = pd.read_sql_query("SELECT * FROM customers;", connection)
print(data)
```
